```python
quar=['Q1','Q2','Q3','Q4']
n=df['Quarter'][0]
n[0:2]
```

```python
df['quar']=0
df['quar']
```

```python
for  i in range(42):
    n=df['Quarter'][i]
    df['quar'][i]=n[0:2]
df['quar']
```

```python
dum=pd.DataFrame(pd.get_dummies(df['quar']))
dum
```

```python
data2=pd.concat((df,dum),axis=1)
data2
```

```python
log_sl=np.log(data2['Sales'])
data2['log_sl']=log_sl
data2
```

```python
t=np.arange(1,43)
data2['t']=t
data2['t_squared']=data2['t']*data2['t']
data2
```

## Spliting Data into train and test

```python
Train =data2.head(84)
Test = data2.tail(12)
```

```python
#Linear Model
import statsmodels.formula.api as smf 

linear_model = smf.ols('Sales~t',data=Train).fit()
pred_linear =  pd.Series(linear_model.predict(pd.DataFrame(Test['t'])))
rmse_linear = np.sqrt(np.mean((np.array(Test['Sales'])-np.array(pred_linear))**2)) #rmse
rmse_linear
```

```python
#Exponential

Exp = smf.ols('log_sl~t',data=Train).fit() #log 
pred_Exp = pd.Series(Exp.predict(pd.DataFrame(Test['t'])))
rmse_Exp = np.sqrt(np.mean((np.array(Test['Sales'])-np.array(np.exp(pred_Exp)))**2))
rmse_Exp
```

```python
#Quadratic 

Quad = smf.ols('Sales~t+t_squared',data=Train).fit()
pred_Quad = pd.Series(Quad.predict(Test[["t","t_squared"]]))
rmse_Quad = np.sqrt(np.mean((np.array(Test['Sales'])-np.array(pred_Quad))**2))
rmse_Quad
```

```python
#Additive seasonality 

add_sea = smf.ols('Sales~quar',data=Train).fit() #y var with dummy vars
pred_add_sea = pd.Series(add_sea.predict(Test[['quar']]))
rmse_add_sea = np.sqrt(np.mean((np.array(Test['Sales'])-np.array(pred_add_sea))**2))
rmse_add_sea
```

```python
#Additive Seasonality Quadratic 

add_sea_Quad = smf.ols('Sales~t+t_squared+quar',data=Train).fit()
pred_add_sea_quad = pd.Series(add_sea_Quad.predict(Test[['quar','t','t_squared']]))
rmse_add_sea_quad = np.sqrt(np.mean((np.array(Test['Sales'])-np.array(pred_add_sea_quad))**2))
rmse_add_sea_quad
```

```python
##Multiplicative Seasonality

Mul_sea = smf.ols('log_sl~quar',data = Train).fit() #log of dummy
pred_Mult_sea = pd.Series(Mul_sea.predict(Test))
rmse_Mult_sea = np.sqrt(np.mean((np.array(Test['Sales'])-np.array(np.exp(pred_Mult_sea)))**2))
rmse_Mult_sea
```

```python
#Multiplicative Additive Seasonality 

Mul_Add_sea = smf.ols('log_sl~t+quar',data = Train).fit()
pred_Mult_add_sea = pd.Series(Mul_Add_sea.predict(Test))
rmse_Mult_add_sea = np.sqrt(np.mean((np.array(Test['Sales'])-np.array(np.exp(pred_Mult_add_sea)))**2))
rmse_Mult_add_sea 
```

```python
#Compare the results 

data3 = {"MODEL":pd.Series(["rmse_linear","rmse_Exp","rmse_Quad","rmse_add_sea","rmse_add_sea_quad","rmse_Mult_sea","rmse_Mult_add_sea"]),"RMSE_Values":pd.Series([rmse_linear,rmse_Exp,rmse_Quad,rmse_add_sea,rmse_add_sea_quad,rmse_Mult_sea,rmse_Mult_add_sea])}
table_rmse=pd.DataFrame(data3)
table_rmse.sort_values(['RMSE_Values'])
```

### from the above results Additive Seasonality Quadratic model gives less rmse value. hence, it is the best model
